```python
import pandas as pa
import seaborn as sns
```

```python
data = pa.read_csv('timings.csv')
data
```

```python
sns.set_theme(style='whitegrid')
g = sns.catplot(data=data[data['op']=='CP2K'], kind='bar', x='nodes', hue='type', y='time max')
g.despine(left=True)
g.set_axis_labels("", "Total time [s]")
g.savefig('cp2k-timings.pdf')
```
